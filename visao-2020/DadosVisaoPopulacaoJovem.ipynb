```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel('tabela1378_correta.xlsx')
```

```python
dataset.head(20)
```

```python
dataset_sexo = pd.DataFrame(columns=['UF','Homens','Mulheres'])
```

```python
#pegando as UFs
listaUF = []
for i in range(len(dataset)):
    nome = dataset.iloc[i,0]
    listaUF.append(nome)
listaUF = [x for x in listaUF if str(x) != 'nan']
listaUF = [x for x in listaUF if str(x) != 'Brasil']
dataset_sexo['UF'] = listaUF
```

```python
dataset_sexo
```

```python
#pegando a quantidade de Homens e somando idades
quantHomens = []
i = 0
max = 112   #len(dataset) = 113
while (i < max):    
    idade = 0
    idade = dataset.iloc[i,3] + dataset.iloc[i+1,3] + dataset.iloc[i+2,3] + dataset.iloc[i+3,3]
    quantHomens.append(int(idade))
    i = i + 4
quantHomens = quantHomens[1:]  #tirando valor do Brasil
dataset_sexo['Homens'] = quantHomens
```

```python
dataset_sexo
```

```python
#pegando a quantidade de Mulheres e somando idades
quantMulheres = []
i = 0
max = 112   #len(dataset) = 113
while (i < max):    
    idade = 0
    idade = dataset.iloc[i,4] + dataset.iloc[i+1,4] + dataset.iloc[i+2,4] + dataset.iloc[i+3,4]
    quantMulheres.append(int(idade))
    i = i + 4
quantMulheres = quantMulheres[1:]
dataset_sexo['Mulheres'] = quantMulheres
```

```python
total_sexo = []
for i in range(len(dataset_sexo)):
    total_sexo.append(dataset_sexo.iloc[i,1] + dataset_sexo.iloc[i,2])
dataset_sexo['Total'] = total_sexo
```
