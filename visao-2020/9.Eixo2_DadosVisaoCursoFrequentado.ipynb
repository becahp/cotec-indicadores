# Parâmetros fixos:

# Função pra gerar o SQL

```python
def gerar_SQL_insert(name_column, group_id, year_id, dataset, id_banco,percentual):
    codigo_sql = []
    for index, row in dataset.iterrows():
        id_banco = id_banco + 1
        region_id = dataset.loc[index,'id']   
        jhi_value = dataset.loc[index,name_column] 
        if percentual:
            jhi_value = round(jhi_value*100,1)
        sql = "INSERT INTO indicator (id,jhi_value,region_id,group_id,year_id)  VALUES (" + str(id_banco) + ",'" + str(jhi_value) + "','" + str(region_id) + "','" + group_id + "','" + year_id + "');"
        codigo_sql.append(sql)
    return codigo_sql
```

```python
# agora posso colocar a coluna ID no dataset... ordem tá correta
dataset2016['id'] = df_auxiliar['id']
     # --> MUDAR ANO DO DATASET
    
dataset2016.head()
```

```python
year_id = "17"      #Ano: 2016
group_id = "449"    # é o primeiro: Alfabetizacao
percentual = False

id_banco = 5154686  # conferir - ultimo valor do ano ou da série histórica anterior
nome_dataset = dataset2016

print('Ano: 2016 - Escolaridade')
print()

for nome in ['Alfabetizacao', 'Alfabetizacao(%)',
       'EnsinoFundamental ', 'EnsinoFundamental(%)', 'EJAEnsinoFundamental',
       'EJAEnsinoFundamental(%)', 'EnsinoMedio', 'EnsinoMedio(%)',
       'EJAEnsinoMedio', 'EJAEnsinoMedio(%)', 'EnsinoSuperior',
       'EnsinoSuperior(%)', 'Especializacao', 'Especializacao(%)', 'Mestrado',
       'Mestrado(%)', 'Doutorado', 'Doutorado(%)', 'Total']:
    print(nome,'\n')
    name_column = nome
    lista = gerar_SQL_insert(name_column, group_id, year_id, nome_dataset, id_banco,percentual)
    for i in lista:
        print(i)
    id_banco = id_banco + 27
    print()
    print('Conferir valor atual do id_banco: ',id_banco)
    print()
    group_id = str(int(group_id)+1) #só no final
    percentual = not(percentual) #só no final tbm
```
