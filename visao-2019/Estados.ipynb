# Limpar strings

Limpar as strings de acentos, cedilhas, apóstrofos e espaços para poder comparar

```python
!pip install Unidecode
import unidecode
```

```python
def clean_string(str):
  str = str.replace(" ","") #remover espaços
  str = str.replace("'","") #remover apóstrofos
  str = str.lower() #colocar tudo em minúsculas
  str = unidecode.unidecode(str) #remover acentuação
  return str
  
```

```python
#exemplo
txt = "Nova Brasilândia D'Oeste oração"
clean_string(txt)
```

# Gerar SQLs de Update

```python
def gerar_SQL(name, name_id, year_id, dataset):
    #name = 'Rural_Total'
    #name_id = "40"     #População Jovem Rural: Total
    #year_id = "9"      #Ano: 2010
    nova_coluna = []
    for index, row in dataset.iterrows():
        region_id = dataset.loc[index,'ID']
        jhi_value = dataset.loc[index,name]

        sql = "UPDATE `indicator` SET `jhi_value`= " + str(jhi_value) + " WHERE `name_id`=" + name_id + " and `year_id`=" + year_id + " and `region_id`=" + str(region_id) + ";"
        nova_coluna.append(sql)
    return nova_coluna
```
