# Limpar strings

Limpar as strings de acentos, cedilhas, apóstrofos e espaços para poder comparar

```python
!pip install Unidecode
import unidecode
```

```python
def clean_string(str):
  str = str.replace(" ","") #remover espaços
  str = str.replace("'","") #remover apóstrofos
  str = str.lower() #colocar tudo em minúsculas
  str = unidecode.unidecode(str) #remover acentuação
  return str
```

```python
#exemplo
txt = "Nova Brasilândia D'Oeste oração"
clean_string(txt)
```

# Novo dataset com nomes limpos

Pegar já os valores que eu quero em um novo dataframe


```python
#nova lista que vai tirar os "(UF)" dos nomes de cidades e copiá-la para os valores nan
novalista = []
for item in dataset['Município'].iteritems(): 
  if type(item[1]) is str:
    mun = item[1]
    if mun.endswith(')'):
      mun = mun[:-4]
      novalista.append(mun)
  else:
    novalista.append(mun)
```

Checar se a nova lista de Municípios é a mesma da original

```python
len(novalista) == len(dataset['Município'])
```

Adiciona nova coluna ao fim do dataset original, agora com todos os nomes de cidades, sem o parênteses de estado

```python
dataset_novo = dataset.assign(editMun= novalista)
```

```python
# Save the result in a new column
dataset_novo['editMun'] = dataset_novo['editMun'].apply(clean_string)
```

```python
dataset_novo.head(10)
```

Agora faço o mesmo para o dataset do visao

```python
dataset_v = dataset_visao.assign(editMun= dataset_visao['Município'])
```

```python
# Save the result in a new column
dataset_v['editMun'] = dataset_v['editMun'].apply(clean_string)
```

```python
dataset_v.head(10)
```

Inserimos o ID do Visao na nossa Tabela do excel


```python
dataset_merge = pd.merge(dataset_novo, dataset_v, how='inner', left_on=['UF','editMun'],right_on=['UF','editMun'])
```
