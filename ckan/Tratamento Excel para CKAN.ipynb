# Tabela de lat/lon

```python
geo_file = '../nova_geo_corrigida.csv'
```

```python
geo = pd.read_csv(geo_file)
```

```python
#pegar apenas estados
estados_geo = geo.loc[geo['capital']==2]
```

# Dados

```python
nomes_excel = []
with os.scandir('../dados/Excel') as entries:
    for entry in entries:
        #print(entry.path)
        print(entry.name)
        nomes_excel.append(entry.path)
```

```python
for nome in nomes_excel:
    #print(nome)
    xls = pd.ExcelFile(nome)
    for ano in xls.sheet_names:
        print(nome,ano)
        tabela = pd.read_excel(nome, sheet_name=ano)
        print(tabela.columns)
        for coluna in tabela.columns:
            if(len(coluna)>30):
                print("!!!!Precisa de ajuste!!!!")
                print()
        
        print()
    print()
```

```python
nomes_atual = []
for nome in nomes_excel:
    if("Eixo1" in str(nome)):
        print(nome)
        nomes_atual.append(nome)
```

```python
for nome in nomes_atual:
    #print(nome)
    xls = pd.ExcelFile(nome)
    for ano in xls.sheet_names:
        print(nome,ano)
        tabela = pd.read_excel(nome, sheet_name=ano)
        print(tabela.columns)
        for coluna in tabela.columns:
            if(len(coluna)>30):
                print("!!!!Precisa de ajuste!!!!")
                print()
        
        print()
    print()
```

```python
# define the name of the directory to be created
base = "../dados/"

for nome in nomes_atual:
    nome_pasta = nome[nome.find("\\")+1:-5]
    path = base + nome_pasta
    try:
        os.mkdir(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
    else:
        print ("Successfully created the directory %s " % path)
```

Ajustes necessários para um conjunto de dados

```python
novas_colunas = ['Sigla', 'UF', 'CarteiraAssinada',
       'CarteiraAssinada(%)', 'CarteiraNaoAssinada',
       'CarteiraNaoAssinada(%)', 'NaoAplicavel',
       'NaoAplicavel(%)', 'Total']
```

```python
for i in novas_colunas:
    if(len(i)>30):
        print("Ainda tem que ajustar!")
```

```python
for nome in nomes_atual:
    #print(nome)
    xls = pd.ExcelFile(nome)
    for ano in xls.sheet_names:
        print("Analisando: ",nome,ano)
        
        #lê a tabela e realiza o merge com as latitudes
        tabela = pd.read_excel(nome, sheet_name=ano)
        
        #nesse caso tem ajuste
        if(nome == "../dados/Excel\Eixo1_CarteiraTrabalho.xlsx"):
            tabela.columns = novas_colunas
            print("Colunas ajustadas!\n")
            print(tabela.columns)
        
        dataset_merge = pd.merge(tabela, estados_geo, how='inner', left_on=['Sigla'],right_on=['uf'])
        dataset_merge.drop(columns=['codigo_ibge','uf', 'nome', 'capital'], inplace=True)
        dataset_merge.rename(columns={"Sigla": "UF", "UF": "Nome"}, inplace=True)
        
        #ajuste de nomes
        nomefile = nome[nome.find("\\")+1:-5]
        base = "../dados/" + nomefile #pasta base
        
        if(ano == "Planilha1"):
            ano = "2010"
            
        filename = str(base)+"/"+str(nomefile)+"_"+str(ano)+'.csv'
        print("Salvando: ", filename)
        dataset_merge.to_csv(filename, index=False, encoding='utf-8')
        print()
```
